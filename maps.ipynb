```python
import time
import requests
from PIL import Image
from io import BytesIO
import os


 

# Define the base URL for the Google Maps Static API
base_url = 'https://maps.googleapis.com/maps/api/staticmap'

 

 

# Define your API key
api_key = 'AIzaSyDQdFNNzAUQx_ee9NpvI_HY5XNjbl1B6-k' # ENTER YOUR API KEY HERE--------------------------

 

 

# Define the size of the images
size = '400x422' # extended for watermark removal, end result will be 400x400

 

 

# Define the zoom level
zoom = '20' # max zoom

 

 

# Define the center point and the lines of longitudes and latitudes
center_lat = 8.452943984738983 # ENTER YOUR STARTING LAT HERE---------------------------------
center_lng = 124.62980050482422 # ENTER YOUR STARTING LONG HERE------------------------------
longitudes = [center_lng + 0.0006 * j for j in range(60)]
latitudes = [center_lat - 0.0006 * i for i in range(50)]

#  8.452943984738983, 124.62980050482422

 

# Check if the directory exists and create it if necessary
# directory = "C:\\Users\\lylet\\Documents\\USJR 3RD YEAR\\Summer\\ProfElec\\Pictures" 

directory = "D:\CLARK\Documents\DOST" 
if not os.path.exists(directory):
    os.makedirs(directory)

 

 


# Iterate over the latitudes and longitudes
for i, lat in enumerate(latitudes):  # Added this line
    for j, lng in enumerate(longitudes):
        # Form the URL for the API request
        url = f'{base_url}?center={lat},{lng}&zoom={zoom}&size={size}&maptype=satellite&style=feature:all|element:labels|visibility:off&key={api_key}'

 

 

        # Send the request
        try:
            response = requests.get(url)

        except:
            continue
 

 

        # Check if the request was successful
        if response.status_code == 200:
            # Save the response as an image
            image = Image.open(BytesIO(response.content))
            image = image.convert('RGB')  # Convert image to RGB mode
            image = image.crop((0, 0, 400, 400))
            lat_s = "{:.4f}".format(lat)
            lng_s = "{:.4f}".format(lng)
            # ENTER YOUR PLACE NAME HERE--------------------------------------------
            image.save(f'{directory}, ({lat_s}, {lng_s}).jpeg', 'JPEG')
        else:
            print('Something went wrong')

  

        # Pause for a short period to avoid making too many requests in a short period of time
        time.sleep(1)
```
